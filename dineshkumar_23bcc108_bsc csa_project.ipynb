```python
import pandas as pd

# Load the dataset
file_name = "Chocolate_Bar_Ratings (1).csv"
df = pd.read_csv (file_name)

# Preview the data
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

# Get the actual column names from the DataFrame
actual_columns = df.columns

# Fill missing values or drop rows/columns
# Check if 'Rating' and 'Company' are in actual columns before filling NaNs
if 'Rating' in actual_columns:
    df['Rating'] = df['Rating'].fillna(df['Rating'].mean())  # Example for numeric column
if 'Company' in actual_columns:
    df['Company'] = df['Company'].fillna('Unknown')  # Example for categorical column
else:
    # If 'Company' is not found, print an error or handle it appropriately
    print("Column 'Company' not found in the DataFrame.")
    # You might want to check for similar column names or adjust your code accordingly
```

```python
# Check for duplicates
print("Number of duplicate rows:", df.duplicated().sum())

# Drop duplicate rows
df = df.drop_duplicates()
```

```python
# Convert 'Cocoa Percentage' to numeric (remove '%' and convert to float)
# Check if the column name exists in the DataFrame before attempting to access it
cocoa_percentage_col = 'Cocoa Percentage'  # Or the actual column name if different

if cocoa_percentage_col in df.columns:
    # Replace '%' and convert to float
    df[cocoa_percentage_col] = df[cocoa_percentage_col].str.replace('%', '').astype(float)
else:
    print(f"Column '{cocoa_percentage_col}' not found in the DataFrame.")
    # Print available columns for debugging
    print("Available columns:", df.columns)

# Convert 'Review Date' to datetime if applicable
# Check if 'Review Date' column exists before conversion
if 'Review Date' in df.columns:
    df['Review Date'] = pd.to_datetime(df['Review Date'], format='%Y')
else:
    print("Column 'Review Date' not found in the DataFrame. Skipping conversion.")
```

```python
# Visualize outliers using boxplots
import seaborn as sns
sns.boxplot(df['Rating'])

# Handle outliers (e.g., capping values)
df['Rating'] = df['Rating'].clip(lower=0, upper=5)  # Example for ratings
```

```python
# Check if 'Company' column exists before processing
if 'Company' in df.columns:
    # Strip whitespace and convert to lowercase for consistency
    df['Company'] = df['Company'].str.strip().str.lower()
else:
    print("Column 'Company' not found in the DataFrame. Skipping this step.")

# Continue with processing 'Bean Origin'
# Check if the column exists before processing
bean_origin_col = 'Bean Origin'  # or the actual column name
if bean_origin_col in df.columns:
    df[bean_origin_col] = df[bean_origin_col].str.strip().str.title()  # Title case
else:
    print(f"Column '{bean_origin_col}' not found in the DataFrame. Skipping this step.")
    # Print available columns for debugging
    print("Available columns:", df.columns)
```

```python
# Save cleaned data to a new CSV file
cleaned_file_name = "Cleaned_Chocolate_Bar_Ratings.csv"
df.to_csv(cleaned_file_name, index=False)

print(f"Cleaned dataset saved as {cleaned_file_name}")
```

```python
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data for the chocolate bar ratings dataset
data = [
    {"Company": "Amedei", "Bean Origin": "Venezuela", "Cocoa Percentage": "70%", "Rating": 4.5, "Review Date": 2022},
    {"Company": "Valrhona", "Bean Origin": "Madagascar", "Cocoa Percentage": "75%", "Rating": 4.2, "Review Date": 2021},
    {"Company": "Lindt", "Bean Origin": "Ecuador", "Cocoa Percentage": "85%", "Rating": 3.8, "Review Date": 2020},
    {"Company": "Green & Black's", "Bean Origin": "Dominican Republic", "Cocoa Percentage": "72%", "Rating": 4.0, "Review Date": 2023},
    {"Company": "Godiva", "Bean Origin": "Peru", "Cocoa Percentage": "65%", "Rating": 3.5, "Review Date": 2021},
]

# Define the file name
file_name = "chocolate_bar_ratings.csv"

# Write data to the CSV file
with open(file_name, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=["Company", "Bean Origin", "Cocoa Percentage", "Rating", "Review Date"])
    writer.writeheader()
    writer.writerows(data)

print(f"Dataset has been written to {file_name}")

# Read the data for visualization
companies = [row["Company"] for row in data]
ratings = [row["Rating"] for row in data]
cocoa_percentage = [int(row["Cocoa Percentage"].replace('%', '')) for row in data]

# 1. Bar Graph - Chocolate Ratings by Company
plt.figure(figsize=(10, 6))
plt.bar(companies, ratings, color="chocolate", alpha=0.8)
plt.title("Chocolate Bar Ratings by Company", fontsize=16)
plt.xlabel("Company", fontsize=12)
plt.ylabel("Rating", fontsize=12)
plt.ylim(0, 5)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2. Heatmap - Cocoa Percentage vs Rating
ratings_matrix = np.array([ratings])  # Reshape ratings into a 2D array for heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(ratings_matrix, annot=True, xticklabels=companies, yticklabels=["Ratings"], cmap="YlGnBu", cbar=True)
plt.title("Heatmap: Cocoa Percentage vs Rating", fontsize=16)
plt.ylabel("Rating", fontsize=12)
plt.xlabel("Company", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 3. Pie Chart - Rating Distribution by Company
plt.figure(figsize=(8, 8))
plt.pie(ratings, labels=companies, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Chocolate Bar Ratings Distribution by Company", fontsize=16)
plt.show()

```

```python
# 4. Scatter Plot - Cocoa Percentage vs Rating
plt.figure(figsize=(10, 6))
plt.scatter(cocoa_percentage, ratings, color="green", s=100)
plt.title("Cocoa Percentage vs Rating", fontsize=16)
plt.xlabel("Cocoa Percentage", fontsize=12)
plt.ylabel("Rating", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for the chocolate bar ratings dataset
data = [
    {"Company": "Amedei", "Bean Origin": "Venezuela", "Cocoa Percentage": "70%", "Rating": 4.5, "Review Date": 2022},
    {"Company": "Valrhona", "Bean Origin": "Madagascar", "Cocoa Percentage": "75%", "Rating": 4.2, "Review Date": 2021},
    {"Company": "Lindt", "Bean Origin": "Ecuador", "Cocoa Percentage": "85%", "Rating": 3.8, "Review Date": 2020},
    {"Company": "Green & Black's", "Bean Origin": "Dominican Republic", "Cocoa Percentage": "72%", "Rating": 4.0, "Review Date": 2023},
    {"Company": "Godiva", "Bean Origin": "Peru", "Cocoa Percentage": "65%", "Rating": 3.5, "Review Date": 2021},
]

# Convert data into a DataFrame
df = pd.DataFrame(data)

# Convert 'Cocoa Percentage' from string to numeric (removing '%' sign)
df['Cocoa Percentage'] = df['Cocoa Percentage'].str.replace('%', '').astype(float)

# Calculate the correlation matrix for numeric columns
correlation_matrix = df[['Cocoa Percentage', 'Rating', 'Review Date']].corr()

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Add a title to the heatmap
plt.title('Correlation Heatmap of Chocolate Bar Ratings')

# Show the plot
plt.show()
```
